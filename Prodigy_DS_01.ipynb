```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_excel(r"C:\Users\sulek\OneDrive\Documents\API_SP.POP.TOTL_DS2_en_excel_v2_42505.xls", header=2)
df
```

```python

# Define the list of selected country names
selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Pakistan']

# Filter data for the year 2022 and selected countries
df_selected = df[df['Country Name'].isin(selected_countries)][['Country Name', '2022']]

# Drop rows with missing values in the '2022' column
df_selected = df_selected.dropna(subset=['2022'])

# Plotting the bar chart for the population of selected countries in 2022
plt.figure(figsize=(12, 8))
plt.bar(df_selected['Country Name'], df_selected['2022'], color='skyblue')

# Set title and labels
plt.title('Population of Selected Countries in 2022')
plt.xlabel('Country')
plt.ylabel('Total Population')

plt.xticks(rotation=45, ha='right')  # Rotate and align country names on x-axis

plt.tight_layout()
plt.show()
```
